```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("Students_Grading_Dataset.csv")
df
```

```python
#.head() to view the first five rows of the updated DataFrame
df.head()
```

```python
#.value_counts() is used to return a Series containing counts of unique values.
df["Department"].value_counts()
```

```python
# Create a mapping dictionary for department abbreviations
Department_dict = {
    "CS": "CS",
    "Engineering": "ENG",
    "Business": "BS",
    "Mathematics": "MATH"
}
```

```python
# Map the department names to their abbreviations
Abbreviated_Department = df["Department"].map(Department_dict)
Abbreviated_Department
```

```python
Abbreviated_Department.value_counts()
```

```python
#
df["Department"] = Abbreviated_Department
df
```

```python
#.head() to view the first five rows of the updated DataFrame
df.head()
```

```python
# Trim the DataFrame to include only relevant columns
selected_columns = [
    "Student_ID",
    "Email",
    "Gender",
    "Age",
    "Department",
    "Attendance (%)",
    "Final_Score",
    "Midterm_Score",
    "Total_Score",
    "Grade",
    "Study_Hours_per_Week",
    "Internet_Access_at_Home",
    "Stress_Level (1-10)",
    "Sleep_Hours_per_Night"
]

trimmed_df = df[selected_columns]
trimmed_df.head()
```

```python
# Attain the attendance column values
Attendance_values = trimmed_df["Attendance (%)"]
Attendance_values
```

```python
#.Describe() is used to generate descriptive statistics that summarize the central tendency, dispersion and shape of a dataset’s distribution, excluding NaN values.
Attendance_values.describe()
```

```python
# Function to classify attendance
#90-100 - EXCELLENT
#50-90 - GOOD
#< 50 - POOR
def Attendance_classifier(attendance):
  if attendance > 90 and attendance <= 100:
    return "Excellent"
  elif attendance > 50 and attendance <= 90:
    return "Good"
  else :
    return "Poor"
```

```python
# Apply the classifier to the attendance values
Classified_Attendance = Attendance_values.apply(Attendance_classifier)
Classified_Attendance
```

#Lambda Functions
- A lambda function in Python is a small, one-line function that does not need a name.
People use it when they need a quick function for a short task.
- Basic structure for the lambda functions is as follows:
lambda arguments: expression

```python
# A list of numbers
numbers = [4,6,5,7,8,10,9,12]
```

```python
#Use map and lambda to square each number in the list
squared_numbers = list(map(lambda x: x**2, numbers))
squared_numbers
```
